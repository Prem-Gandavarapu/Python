```python
n = int(input())

for i in range(2 * n - 1):
    if i < n:
        stars = i + 1
    else:
        stars = 2 * n - i - 1

    for j in range(stars):
        print("*", end=" ")
    print()
```

```python
n = int(input())
for i in range(n):
    if i % 2 == 0:
        start = 1
    else:
        start = 0
    for j in range(i + 1):
        print(start, end=" ")
        start = 1 - start
    print()
```
